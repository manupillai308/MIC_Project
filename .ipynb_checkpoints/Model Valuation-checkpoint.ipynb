```python
import cv2
import numpy as np

def draw_mask(image, mask_generated, color) :
    if image.dtype != np.uint8:
        image = (image.copy()*255).astype(np.uint8)
    if len(image.shape) < 3:
        image = cv2.cvtColor(image, cv2.COLOR_GRAY2RGB)
    masked_image = image.copy()

    masked_image = np.where(mask_generated.astype(int),
                          np.array(color, dtype='uint8'),
                          masked_image)

    masked_image = masked_image.astype(np.uint8)
    
    return masked_image
```

```python
RED = [255, 0, 0]
GREEN = [0, 255, 0]
BLUE = [0, 0, 255]
```

```python
def dilate(label, k=15, i=10):
    return cv2.dilate(label.unsqueeze(2).numpy().astype(np.uint8), (k, k), iterations=i).reshape(240, 240, 1)

def erode(label, k=15, i=10):
    return cv2.erode(label.unsqueeze(2).numpy().astype(np.uint8), (k, k), iterations=i).reshape(240, 240, 1)
```
